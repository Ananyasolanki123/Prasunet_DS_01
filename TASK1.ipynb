```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("/content/Earthquake Perception Dataset.csv")
```

```python
df
```

```python
df1=df.isnull().sum()
df1
```

```python
df1 = df.fillna(method='ffill')
df1 = df.fillna(method='bfill')
df2=df1

df2
```

```python
df2.isnull().sum()
```

```python
df_cleaned=df2.dropna(subset=['political_orientation','salary'])
```

```python
df_cleaned
```

```python
df_cleaned.info()
```

```python
df_cleaned['age'] = df_cleaned['age'].astype(int)
```

```python
df_cleaned.duplicated().sum()
```

```python
sns.violinplot(x='age', y='sex', data=df_cleaned)
plt.show()
```

```python
sns.lineplot(x='age', y='physiological_symptoms', data=df_cleaned)
plt.show()
```

```python

contingency_table = pd.crosstab(df_cleaned['earthquake80'], df_cleaned['red_zone_frequency'])
sns.heatmap(contingency_table, annot=True, cmap='YlGnBu')
plt.show()
```
